```python
import pandas as pd
import numpy as np
```

```python
# veri üzerinde standart sapma mod medyan ve boş değerlerin doldurulması  ve satır ve sutunlar üzerinde bazı işlemler 
```

```python
dataset = pd.read_excel("diyabet.XLSX")  # veri setini yükleme
```

```python
dataset  #veri setinin bütün satır ve sütün değerlerini ekrana yazdırma
```

```python
print("Satır sayısı =",len(dataset.index)) # satır sayısı
print("Sütun sayısı =",len(dataset.columns)) # sütun sayısı 
```

```python
liste = ["'preg'","'plas'","'skin'","'insu'","'mass'","'pedi'","'age'"]
```

```python
#Burda veri setinin sütun isimlerini önce bir listeye aktardım daha sonra kolaylık olsun 
#diye Her sütunun ortalaması, minimum değeri, maksimum değeri, standart sapması nı bir fonksiyonla yazdırdım
```

```python
def deger_hesapla():
    for i in liste:
        print(i," Sutünuna ait Max, Min, Ortalama ve Standart Sapma değerleri ")
        print("-"*45)
        print("Max sonucu = ",dataset[i].max())
        print("Ortalama sonucu = ",dataset[i].mean())
        print("Min sonucu = ",dataset[i].min())
        print("Standart sapma sonucu = ",dataset[i].std())
        print(" "*45)
       
        
```

```python
deger_hesapla()
```

```python
# burda her satırın max min ortalama ve standart sapma hesaplamalarını bir  fonksiyonla yazdırdım
```

```python
def hesapla():
    for i in range(len(dataset.index)):
        arr =np.array(dataset.loc[[i],["'preg'","'plas'","'plas'","'skin'","'insu'","'mass'","'pedi'","'age'"]])
        print(i,". Satırın Max, Min, Ortalama ve Standart Sapma Değerleri ")
        print(" Max değeri =",arr.max())
        print("Min değeri =",arr.min()) 
        print("Ortalama değeri =",arr.mean())
        print("Standart sapma değeri",arr.std())
        print("-"*45)
        
```

```python
hesapla()
```
