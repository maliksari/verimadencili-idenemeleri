```python
# veri üzerinde bazı sınıflandırmalar ve veri görselleştirme 

import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score 
from sklearn.cluster import AgglomerativeClustering 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA 
from sklearn.metrics import silhouette_score 
import scipy.cluster.hierarchy as shc
from sklearn.preprocessing import StandardScaler, normalize 
```

```python
data = pd.read_excel("data.xlsx")
```

```python
# boş değerlerin ortalama ile doldurulması 
data['X'] = data['X'].fillna(data['X'].mean()).astype(int)
data['Y'] = data['Y'].fillna(data['Y'].mean()).astype(int)
```

```python
def k_means():
    #k-means  k = 2 için
    
    model = KMeans(n_clusters=2)
    model.fit(data)
    y_kmeans = model.predict(data)
    plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=y_kmeans, s=50, cmap='rainbow')
    
    centers = model.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], c='green', s=200,marker ='s');
    plt.title('n_clusters =2 için Grafik')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    
    # k means k = 4 
    model = KMeans(n_clusters=4)
    model.fit(data)
    KMeans(algorithm='auto', copy_x=True, init='k-means++', max_iter=300,
       n_clusters=4, n_init=10, n_jobs=1, precompute_distances='auto',
       random_state=None, tol=0.0001, verbose=0)
    y_kmeans = model.predict(data)
    plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=y_kmeans, s=100, cmap='rainbow')
    centers = model.cluster_centers_
    plt.scatter(centers[:, 0], centers[:, 1], c='green', s=200,marker ='s');
    plt.title('n_clusters =4 için Grafik')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
    
```

```python
k_means()
```

```python
def agglomerative():
    # agglomerative yöntemi ile sınıflandırma
    ac2 = AgglomerativeClustering(n_clusters = 3) 
   
    plt.figure(figsize =(6, 6)) 
    plt.scatter(data['X'], data['Y'],  
           c = ac2.fit_predict(data),s=100, cmap ='rainbow') 
    plt.show() 
```

```python
agglomerative()
```

```python
def expectation():
    gmm = GaussianMixture(n_components=3)
    gmm.fit(data)
    plt.scatter(data['X'], data['Y'],  
           c = gmm.fit_predict(data),s=100, cmap ='viridis')
 
    plt.show()
```

```python
expectation()
```
