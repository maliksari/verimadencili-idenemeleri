```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import KFold
```

```python
dataset = pd.read_excel("diyabet.XLSX",index_col = 'id') 
label = LabelEncoder() 
dataset['class'] = label.fit_transform(dataset['class'])  # negatif değerlerin 0 a pozitif değerlerin 1 e dönüşümü
```

```python
# k kat çaprazlama k= 5 alınmıştır
def K_kat():
    X = np.array(dataset.drop(['class'], 1).astype(float))
    y = np.array(dataset['class'])
    
    kf = KFold(n_splits=5) # k = 5
    kf.get_n_splits(X)
    for train_index, test_index in kf.split(X):
        print("TRAIN:", train_index, "TEST:", test_index)
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
    
```

```python
K_kat()
```

```python
# K means sınıflandırması
def K_en():
    a = np.array(dataset.drop(['class'], 1).astype(float))
    y = np.array(dataset['class'])
    kmeans = KMeans(n_clusters=5) # K = 5 ile sınıflandırma
    kmeans.fit(a)
    correct = 0
    for i in range(len(a)):
        tahmin = np.array(a[i].astype(float))
        tahmin = tahmin.reshape(-1, len(tahmin))
        T = kmeans.predict(tahmin)
        if T[0] == y[i]:
             correct += 1

    print(correct/len(a))
```

```python
K_en()
```

```python
a = dataset[["preg","plas","pres","skin","insu","mass","pedi","age"]]
b = dataset["class"] 
X_train, X_test, y_train, y_test = train_test_split(a, b, test_size=0.3)   #verilerin %30 test verisi olarak ayrılması
```

```python
# destek vektör makinesi ile sınıflandırma 
def dvm():
    
    clf = svm.SVC(kernel='linear') 
   
    clf.fit(X_train, y_train)
 
    y_pred = clf.predict(X_test)
    print("Başarı:",metrics.accuracy_score(y_test, y_pred))
    
```

```python
dvm()
```

```python
def bagging():
    # değerlerin büyük olmasından öncelikle min-max normalizasyonu uyguladım daha sonra bagging sınıflandırıcısını kullandım
    
    deger = dataset.values

    imputer = Imputer()
    imputedData = imputer.fit_transform(deger)
    scaler = MinMaxScaler(feature_range=(0, 1))
    
    normalizedData = scaler.fit_transform(imputedData)
    
    X = normalizedData[:,0:8]
    Y = normalizedData[:,8]
    
    kfold = model_selection.KFold(n_splits=10, random_state=1)
    cart = DecisionTreeClassifier()
    num_trees = 100
    model = BaggingClassifier(base_estimator=cart, n_estimators=num_trees, random_state=1)
    results = model_selection.cross_val_score(model, X, Y, cv=kfold)
    print(results.mean())
    
```

```python
bagging()
```

```python
# imputerde oluşan hata sürümle alaklı bi durum 
```

```python
# boosting ile sınıflandırma
def boosting():
    abc = AdaBoostClassifier(n_estimators=50,learning_rate=1)

    model = abc.fit(X_train, y_train)


    y_pred = model.predict(X_test)
    print("Başarı:",metrics.accuracy_score(y_test, y_pred))
```

```python
boosting()
```
